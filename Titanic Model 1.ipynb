```python
import pandas as pd
import numpy as np

train = pd.read_csv("D:/Users/taharam/Desktop/Kaggle/Titanic/train.csv")
test = pd.read_csv("D:/Users/taharam/Desktop/Kaggle/Titanic/test.csv")
```

```python
def misdata_table(df):
    #how many missing data are there?
    null_val = df.isnull().sum()
    #the percentage of having missing data
    percent = 100*df.isnull().sum()/len(df)
    #combining data / creating 2 columns
    misdata_table = pd.concat([null_val, percent], axis=1)
    misdata_table_ren_columns = misdata_table.rename(columns = {0: 'missing data', 1:'%'})
    return  misdata_table_ren_columns 
```

```python
misdata_table(train)
```

```python
misdata_table(test)
misdata_table(train)
```

```python
#Filling out the missing data by median 
train['Age'] = train['Age'].fillna(train['Age'].median())
train['Embarked'] = train['Embarked'].fillna('S')
```

```python
#Converting str data to int data
train['Sex'][train['Sex'] == 'male'] = 0
train['Sex'][train['Sex'] == 'female'] = 1
train['Embarked'][train['Embarked'] == 'S'] = 0
train['Embarked'][train['Embarked'] == 'C'] = 1
train['Embarked'][train['Embarked'] == 'Q'] = 2
train.head(10)
```

```python
#Same thing for test data
test['Age']=test['Age'].fillna(test['Age'].median())
test['Sex'][test['Sex'] == 'male'] = 0
test['Sex'][test['Sex'] == 'female'] = 1
test['Embarked'][test['Embarked'] == 'S'] = 0
test['Embarked'][test['Embarked'] == 'C'] = 1
test['Embarked'][test['Embarked'] == 'Q'] = 2
test.Fare[152] = test.Fare.median()
test.head(10)
```

```python
#Prediction model; scikit-learn / Numpy
from sklearn import tree
```

```python
#1st Model; Considering Pclass, Sex, Age, Fare
#Getting the explained variables of train
target = train['Survived'].values
features_one = train[['Pclass', 'Sex','Age','Fare']].values
#Making decision trees by using scikit-learn
my_tree_one = tree.DecisionTreeClassifier()
my_tree_one = my_tree_one.fit(features_one, target)
```

```python
#Getting the explained variables of test
test_features = test[['Pclass','Sex','Age','Fare']].values
```

```python
#Predict the model by using explained variables
my_prediction = my_tree_one.predict(test_features)
```
